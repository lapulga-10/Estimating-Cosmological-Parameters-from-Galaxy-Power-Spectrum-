# 2) Define Simulator to Generate the data.

```python
def get_noisy_pk(H0, Omega_m, n_s, k_vals=None, noise_level=0.2, seed=None):
    """
    Generate noisy matter power spectrum P(k) from given cosmological parameters.

    Parameters:
        H0 (float): Hubble constant (km/s/Mpc)
        Omega_m (float): total matter density (Î©_m)
        n_s (float): spectral index
        k_vals (np.array or None): k values (1/Mpc). Default: logspace from 1e-3 to ~3.16
        noise_level (float): relative Gaussian noise (e.g., 0.05 = 5%)

    Returns:
        noisy_P_k (np.array): noisy matter power spectrum at redshift z=0
        k_vals (np.array): corresponding k-values
    """
    # Default k range: safe values for CAMB
    if k_vals is None:
        k_vals = np.logspace(-4, 0.5, 1000)

    ombh2 = 0.022
    h = H0 / 100.0
    omch2 = Omega_m * h**2 - ombh2

    # Sanity check to avoid invalid input
    #if omch2 <= 0:
        #raise ValueError(f"Invalid parameters: omch2={omch2} is non-positive.")

    # Setup CAMB
    pars = camb.CAMBparams()
    pars.set_cosmology(H0=H0, ombh2=ombh2, omch2=omch2)
    pars.InitPower.set_params(ns=n_s)
    pars.set_matter_power(redshifts=[0], kmax=max(k_vals))

    # Run CAMB and get P(k)
    PK = camb.get_matter_power_interpolator(pars, nonlinear=False, kmax=max(k_vals))
    P_k = PK.P(0, k_vals)

     # Conditional RNG
    if seed is not None:
        rng = np.random.default_rng(seed)  # reproducible
    else:
        rng = np.random.default_rng()      # fresh random

    # Add Gaussian noise
    noise = rng.normal(0, noise_level * P_k)
    noisy_P_k = P_k + noise

    return noisy_P_k, k_vals
```

```python
H0 = 70
Omega_m = 0.3
n_s = 0.96

P_k_noisy, k_vals = get_noisy_pk(H0, Omega_m, n_s, seed = 42)

print("P(k) shape:", P_k_noisy.shape)
print("NaNs in output?", np.isnan(P_k_noisy).any())
```

# a) Simulator for $log_{10}(P(k))$

```python
class Simulator(bf.simulators.Simulator):
    def __init__(self, k_vals=None, noise_level=1):
        self.noise_level = noise_level
        self.k_vals = k_vals if k_vals is not None else np.linspace(-3, 2, 1024)

    def sample_prior(self, batch_size):
        H0 = np.random.uniform(40,90, size=(batch_size, 1))        # Hubble constant
        Omega_m = np.random.uniform(0.25, 0.35, size=(batch_size, 1))  # Matter density
        n_s = np.random.uniform(0.9, 1.0, size=(batch_size, 1))     # Spectral index
        theta = np.hstack([H0, Omega_m, n_s])
        return theta

    def generate_pk(self, H0, Omega_m, n_s):
        pars = camb.CAMBparams()
        ombh2 = 0.022  # Fixed baryon density
        h = H0 / 100.0
        omch2 = Omega_m * h**2 - ombh2  # Ensure total matter = Omega_m

        pars.set_cosmology(H0=H0, ombh2=ombh2, omch2=omch2)
        pars.InitPower.set_params(ns=n_s)
        pars.set_matter_power(redshifts=[0], kmax=max(self.k_vals))
        PK = camb.get_matter_power_interpolator(pars, nonlinear=False, kmax=max(self.k_vals))
        return PK.P(0, self.k_vals)

    def sample(self, batch_size):
        theta_batch = self.sample_prior(batch_size)
        log_Pk_batch = []

        for i in range(batch_size):
            H0, Om_m, ns = theta_batch[i]
            pk = self.generate_pk(H0, Om_m, ns)
            
            log_P_k = np.log10(pk)
            
            # Add noise to simulate observations
            log_P_k = log_P_k + np.random.normal(loc=0.0, scale=1, size=pk.shape)

            log_Pk_batch.append(log_P_k)

        log_Pk_batch = np.array(log_Pk_batch)
        return dict(theta=theta_batch, log_Pk=log_Pk_batch, P_k = pk, kvals = k_vals)
```
