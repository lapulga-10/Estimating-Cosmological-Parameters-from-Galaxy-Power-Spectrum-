# Data generating process

## Definitions

```python
def generate_log10PK(H0, Omega_m, n_s, k_vals=None, noise_level=0.2, seed=None):
    """
    Generate noisy matter power spectrum P(k) from given cosmological parameters.

    Parameters:
        H0 (float): Hubble constant (km/s/Mpc)
        Omega_m (float): total matter density (Î©_m)
        n_s (float): spectral index
        k_vals (np.array or None): k values (1/Mpc). Default: logspace from 1e-3 to ~3.16
        noise_level (float): relative Gaussian noise (e.g., 0.05 = 5%)

    Returns:
        noisy_P_k (np.array): noisy matter power spectrum at redshift z=0
    """
    # Default k range: safe values for CAMB
    if k_vals is None:
        k_vals = np.logspace(-4, 4, 1024)


    ## To calculate the physical densities from the densities 
    # Baryon density
    Omega_b = 0.05
    
    # Cold dark matter density
    Omega_c = Omega_m - Omega_b

    # dimensionless Hubble parameter
    h = H0 / 100

    
    # physical densities
    ombh2 = Omega_b * h**2
    omch2 = Omega_c * h**2


    # Setup CAMB
    pars = camb.CAMBparams()
    pars.set_cosmology(H0=H0, ombh2=ombh2, omch2=omch2)
    pars.InitPower.set_params(ns=n_s)
    pars.set_matter_power(redshifts=[0], kmax=max(k_vals))

    # Run CAMB and get P(k)
    PK = camb.get_matter_power_interpolator(pars, nonlinear=False, kmax=max(k_vals))
    P_k = PK.P(0, k_vals)

     # Conditional RNG
    if seed is not None:
        rng = np.random.default_rng(seed)  # reproducible
    else:
        rng = np.random.default_rng()      # fresh random

    # Add heteroscedastic Gaussian noise.
    # Noise is added according to the scale of P_k which make more realistic observations
    noise = rng.normal(0, noise_level * P_k)
    P_k_noisy = P_k + noise
    
    
    #Log_Pk_noisy = np.log10(P_k + noise)
    #Log_Pk = np.log10(P_k)

    #Log_kvals = np.log10(k_vals)

    return {
        'k' : k_vals,
        'P_k_clean' : P_k, 
        'P_k_noisy' : P_k_noisy
        #,
        #'Log_k': Log_kvals,
        #'Log_Pk_noisy': Log_Pk_noisy,
        #'Log_Pk_clean': Log_Pk
    }
```

## Definitions part 2

Sample from prior: $(\theta^{(s)}) \sim P(\theta)$

```python
def prior(seed=None):
    rng = np.random.default_rng(seed)

    H0 = rng.uniform(30, 100)
    Omega_m = rng.uniform(0.1, 0.5)
    n_s = rng.uniform(0.7, 1.0)

    return {
        'H0': H0,
        'Omega_m': Omega_m,
        'n_s': n_s
    }
```

```python
def prior_wide(seed=None):
    rng = np.random.default_rng(seed)



    # --- H0: Scaled Beta(2.5, 2.0) over [10, 100]
    a, b = 10, 100
    alpha, beta_param = 2.5, 2.0
    u = rng.beta(alpha, beta_param)  # u â [0, 1]
    H0 = a + (b - a) * u                              # scale to [10, 100]


    
    Omega_m = rng.normal(loc=0.5, scale= 1)
    n_s = rng.normal(loc=0.5, scale=1)





    return {
        'H0': H0,
        'Omega_m': Omega_m,
        'n_s': n_s
    }
```

```python
import numpy as np
from scipy.stats import beta

def beta_parameters_from_mean_variance(mean, variance):
    if not (0 < mean < 1):
        raise ValueError("Mean must be between 0 and 1 for Beta distribution.")
    if variance <= 0 or variance >= mean * (1 - mean):
        raise ValueError("Variance must be > 0 and < Î¼(1-Î¼) for valid Beta.")
    
    temp = (mean * (1 - mean)) / variance - 1
    alpha = mean * temp
    beta_param = (1 - mean) * temp
    return alpha, beta_param

def sample_H0_prior(mean, variance, h0_min=10, h0_max=100, size=1):
    """
    Samples from a scaled Beta prior for the Hubble constant H0.

    Parameters:
        mean (float): Desired mean of H0 prior (must be between h0_min and h0_max).
        variance (float): Desired variance of H0 prior.
        h0_min (float): Lower bound of H0 prior support.
        h0_max (float): Upper bound of H0 prior support.
        size (int): Number of samples to draw.

    Returns:
        np.ndarray: Samples from the prior.
    """
    # Convert to [0, 1] range
    mean_std = (mean - h0_min) / (h0_max - h0_min)
    var_std = variance / (h0_max - h0_min)**2

    # Get alpha and beta
    alpha, beta_param = beta_parameters_from_mean_variance(mean_std, var_std)

    # Sample in [0, 1], then scale to [h0_min, h0_max]
    samples = h0_min + (h0_max - h0_min) * np.random.beta(alpha, beta_param, size=size)
    return samples
```

```python
samples = sample_H0_prior(mean=70, variance=25, h0_min=50, h0_max=90, size=1000)

import matplotlib.pyplot as plt
plt.hist(samples, bins=50, density=True)
plt.title("Samples from Scaled Beta Prior for $H_0$")
plt.xlabel("$H_0$ [km/s/Mpc]")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

```python
# Parameters
a, b = 10, 100
alpha, beta_param = 2, 1.5  # Flatter shape, same mean

# x-values in [10, 100]
x_vals = np.linspace(a, b, 500)

# True PDF of scaled Beta
scaled_beta_pdf = (1 / (b - a)) * beta.pdf((x_vals - a) / (b - a), alpha, beta_param)

# Plot
plt.plot(x_vals, scaled_beta_pdf, linewidth=2)
plt.title(f'Flatter PDF: Scaled Beta({alpha}, {beta_param}) over [{a}, {b}]')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

Sample from the Likelihood: $(y^{(s)}) \sim P(y | \theta)$
The same function but with just one output of length of k_vals

```python
def likelihood(H0, Omega_m, n_s, k_vals=None, noise_level=0.2, seed=None):
    """
        Same function as generate_PK(), the only difference is the single output "P_K_noisy".
        Redefining, just make it Likelihood.
        Also, note that n = len(k_vals) = 128
    """
    # Default k range: safe values for CAMB
    if k_vals is None:
        k_vals = np.logspace(-4, 0.5, 1024) 


    ## To calculate the physical densities from the densities 
    
    # Baryon density (Well Know Constant from Plank 2018)
    Omega_b = 0.05
    
    # Cold dark matter density
    Omega_c = Omega_m - Omega_b

    # dimensionless Hubble parameter
    h = H0 / 100

    
    # physical densities
    ombh2 = Omega_b * h**2
    omch2 = Omega_c * h**2


    # Setup CAMB
    pars = camb.CAMBparams()
    pars.set_cosmology(H0=H0, ombh2=ombh2, omch2=omch2)
    pars.InitPower.set_params(ns=n_s)
    pars.set_matter_power(redshifts=[0], kmax=max(k_vals))

    # Run CAMB and get P(k)
    PK = camb.get_matter_power_interpolator(pars, nonlinear=False, kmax=max(k_vals))
    P_k = PK.P(0, k_vals)

     # Conditional RNG
    if seed is not None:
        rng = np.random.default_rng(seed)  # reproducible
    else:
        rng = np.random.default_rng()      # fresh random

    # Add heteroscedastic Gaussian noise.
    # Noise is added according to the scale of P_k which make more realistic observations
    noise = rng.normal(0, noise_level * P_k)
    P_k_noisy = P_k + noise
    P_k_noisy = P_k_noisy.reshape(-1, 1)

    return {'P_k_noisy': P_k_noisy}
```
